## Reading Images

```python
img = cv2.imread('../data/00-puppy.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

* Matplotlib expects image channels in: R G B order
* OpenCV expects image channels in: B G R order

So openCV is reading image in BGR order whereas matplotlib is displaying it in RGB order...R and B channels are swapped. This is the reason for discolouration.

```python
# color conversion
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

## Flipping images

```python
flip_hori_img = cv2.flip(fix_img,0)
plt.imshow(flip_hori_img)
```

* 0 = flip along horizontal axis
* 1 = flip along vertical axis

```python
flip_vertical_img  =cv2.flip(fix_img,1)
plt.imshow(flip_vertical_img)
```

```python
# Both flips at once
flip_img = cv2.flip(fix_img,-1)
plt.imshow(flip_img)
```

## Drawing on images - Basic shapes

```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
# Draw a rectangle on the image
# pt1= top left corner coordinates
# pt2= bottom right corner coordinates 
cv2.rectangle(blank_img, pt1=(210,100), pt2=(435,250), color=(0,255,0), thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
# Draw a square
cv2.rectangle(blank_img, pt1=(200,200), pt2=(300,300), color=(0,0,255), thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
# draw a circle
cv2.circle(img=blank_img, center=(100,100), radius=50, color= (255,0,0), thickness=8)
plt.imshow(blank_img)
```

```python
# draw another circle
cv2.circle(img=blank_img, center=(400,400), radius=50, color=(255,0,0), thickness=-1)  # thickness=-1 fils the circle
plt.imshow(blank_img)
```

```python
# draw a line
cv2.line(blank_img, pt1=(0,0), pt2=(512,512), color=(102,255,255), thickness=5)
plt.imshow(blank_img)
```
